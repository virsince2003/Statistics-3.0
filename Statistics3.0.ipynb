Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import numpy as np
import matplotlib.pyplot as plt
def normal_distribution(mean,std_dev,x):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-((x - mean)**2) / (2 * std_dev**2))
```

```python
mean = 2  
std_dev = 0.5 
pdf_at_x = normal_distribution(mean, std_dev, x) 
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution PDF')
plt.legend("PDF")
plt.plot(x, pdf_at_x,label='PDF')
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

    The Binomial distribution deals with a fixed number of independent trials (n).
    Each trial has only two possible outcomes: success (usually labeled as 1) or failure (usually labeled as 0).
    The probability of success in each trial is denoted by 'p', and the probability of failure is '1 - p'.
    The distribution is discrete, as it deals with a count of successes in 'n' trials.

Examples:

    Customer Conversion Rate: In marketing, you might have a conversion rate 'p' representing the probability of a customer making a purchase after visiting your website. If 'n' customers visit your website, the number of successful conversions (purchases) among them follows a Binomial distribution.

    Quality Control in Manufacturing: Consider a manufacturing process where each item produced has a probability 'p' of being defective. If 'n' items are produced, the number of defective items among them follows a Binomial distribution. This is commonly used in quality control to assess the proportion of defective items in a batch.

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
import math
def poision_dis(n,mean):
    x=range(0,n+1)
    cdf=0.0
    for x in range(n + 1):
        cdf=+math.exp(-mean) * (mean**x) /math.factorial(x)
    return cdf
```

Q7. How Binomial distribution different from Poisson distribution?

Binomial Distribution:

    Fixed number of independent trials.
    Two possible outcomes (success/failure).
    Parameters: Probability of success (p) and number of trials (n).
    Used for binary events with a fixed number of trials.
    
Poisson Distribution:

    Count of events in a fixed interval.
    Counts range from 0 to positive infinity.
    Parameter: Average rate of occurrence ().
    Used for modeling rare events or events occurring randomly in time or space.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

    Data points close to the mean have a higher frequency of occurrence.
    As we move away from the mean towards the tails of the distribution, the frequency of data points decreases.
    So, the data points in the tails, which are farther from the mean, are less frequent and represent extreme or rare events in the context of the distribution.
